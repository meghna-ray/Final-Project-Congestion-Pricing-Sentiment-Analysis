#Gauging public opinion regarding the introduction of congestion pricing policy in Manhattan, New York through text analysis. 

##Introduction

New York City is looking to reduce emissions, tackle congestion and increase investment in transit. [(NYT, Aug 2022)](https://www.nytimes.com/2022/08/18/nyregion/nyc-congestion-pricing-manhattan.html) It plans to achieve these 3 goals by introducing congestion pricing in the Manhattan region. The congestion pricing policy is a planned pricing scheme that will charge vehicles travelling into or within the central business districts of Manhattan. Prices would vary depending on what type of vehicle is entering the district and at what time. This policy will most likely be implemented by the second quarter of 2024 [(New York City Council, Feb 2023)](https://council.nyc.gov/joseph-borelli/2023/02/27/nyc-congestion-pricing-pushed-back-to-2024-costing-250m-in-revenue-mta-says/#:~:text=Rossville%2C%20Tottenville%2C%20Woodrow-,NYC%20congestion%20pricing%20pushed%20back%20to%202024%2C%20costing,M%20in%20revenue%2C%20MTA%20says&text=Published%3A%20Feb.,%2C%202023%2C%203%3A35%20p.m.) and is the first of its kind in the United States.[(NYT, Aug 2022)](https://www.nytimes.com/2022/08/26/nyregion/congestion-pricing-nyc-mta.html#:~:text=The%20tolling%20plan%2C%20known%20as,raising%20money%20for%20public%20transit.). This policy has faced equal amount of support and opposition from local stakeholders including local citizens, transit officials, the city and state government, as well as those from the surrounding regions of New York such as New Jersey. 
This research project seeks to gauge public opinion regarding the introduction of congestion pricing by conducting a text analysis of news coverage by local publications on the topic.  

Research Question: What is the sentiment of the local population on the proposed introduction of congestion pricing in New York, to reduce traffic in the Manhattan? 

Hypothesis: There is a mixed opinion regarding the introduction of congestion pricing, and there is most likely a relationship between the ideological leanings of the newspaper publication and its coverage of the topic. 

The project at hand seeks to run a preliminary text analysis on the coverage of congestion pricing by local news publications. Firstly, a sentiment analysis of the articles published by local publications on the topic will be conducted. This will help determine the presence of a link between ideological leanings and the tone of the articles, i.e., a positive or negative tone, published by them. This tone will be determined by the sentiment scores assigned to the articles. In addition, the frequency of words used by publications in their coverage of the issue will be analyzed to determine the presence of a link between the publication, the ideological leaning and the issues related to congestion pricing covered by them. 

Running a sentiment analysis will help provide key insights to the public opinion on the matter of congestion pricing. This will provide policy makers a better understanding of the potential concerns and possible obstacles that could surround this policy. For instance, if the public opinion is overwhelmingly negative, policy makers can take appropriate steps to address the negative opinions surrounding it. Policymakers would also be in a better position to make informed decisions regarding the design and implementation plan of the policy. The ability to gain insights from diverse perspectives, will provide them with a more holistic understanding of the concerned policy. In terms of future application of this research, running a sentiment analysis can also help analyze the effectiveness of the policy after its implementation. Running such an analysis after implementation will help policymakers gain a better understanding of the hurdles being faced by the public. It can also help gauge if the sentiment towards the policy has changed post its implementation. Following such analysis, appropriate corrective action could be taken by policy makers. 

##Research Methodology

In order to carry out the above-mentioned analysis, the following steps were undertaken:

###A.	Data collection

For the purposes of this project 30 articles from 3 local news publications were ultimately analyzed. The data collection process included the following steps:
i.	Selection of Articles 
ii.	Web scraping

####i.	Selection of Articles

While building the repository of articles to be selected for the purposes of this project the following criteria were taken into consideration: 
a.	Local News Publication: Although congestion pricing has been part of a national conversation, it is ultimately a local policy issue. Therefore, to gain more authentic commentary on the issue, only local news publications were selected. 

b.	Frequency of coverage: Only those publications that had published at least 10 articles covering the topic of congestion pricing in the past one year, i.e., the period between April 2022 and April 2023 were selected. 

c.	Political ideology: To ensure that the articles selected were representative of diverse ideologies, newspaper publications with diverse political leanings were identified. The ideological leanings of publications were determined through the ratings provided by three websites that analyze media bias. These were: 
1.	[Allsides](https://www.allsides.com/unbiased-balanced-news) 
2.	[Media Bias Fact Check ](https://mediabiasfactcheck.com)
3.	[Biasly](https://www.biasly.com)

Based on an analysis of the media bias provided by these websites, 4 newspaper publications were shortlisted in the initial stage. The selected publications along with political leanings determined for the purposes of this project are as follows: 

1.	The New York Times-Left
2.	New York Daily-Left
3.	Gothamist-Centre-Left
4.	New York Post-Right

Due to the absence of a rating on New York Daily by Biasly, ratings provided by All Sides and Media Bias Fact Check were ultimately used to determine the political leaning for the purposes of this project. A detailed analysis of the media bias ratings can be found in [Appendix 1](https://docs.google.com/document/d/12yI2R7XC5axk46oBP2pRppmxmk_fu7qhcoLpJJynVxg/edit?usp=share_link). Given that New York city in general is a Democratic State, it was difficult to find any other local publications with a right leaning stance, that met all the other stated selection criteria.

d.	Whether the websites of the publications allowed web scraping. There were no restrictions placed on the articles selected from three of the four selected publications i.e the New York Post, Gothamis and New York Daily. 
However, restrictions on web scraping had been placed by the New York Times. While a separate section of code had been written to web scrape articles from the New York Times using an API key, only the headline, and lead paragraph of the articles could be scraped. The absence of complete articles were likely to cause inaccurate results. Therefore, although 41 articles were initially selected to be included in the repository for analysis, articles from the New York times ultimately had to be excluded. The section of code developed to web scrape articles from the New York Times can be found [here](https://github.com/meghna-ray/Final-Project-Congestion-Pricing-Sentiment-Analysis/blob/main/NYT%20Webscraping-Meghna.ipynb). The list of articles selected can be found [here](https://github.com/meghna-ray/Final-Project-Congestion-Pricing-Sentiment-Analysis/blob/main/NYT%20Articles.csv).

####ii.	Web Scraping: 
Once the list of 30 articles were selected a database was made in the form of a .csv file.[(Appendix 2)](https://drive.google.com/file/d/1dZbrILigZtsXbwGNUuGdzVhABWuEjhOT/view?usp=share_link) This .csv file was imported as a pandas data frame to enable analysis. These articles were then web scraped through requests, and the results of the web scraping for each article was then added into a separate column in the data frame. 

###B.	Data Cleaning: 
Once the articles and the web scraped content was added onto the data frame, the data was cleaned:  The following steps were undertaken as a part of this process: 

####i.	Extraction of article body and headline: 
Since the initial process of webscraping, extracted the entire HTML code, beautiful soup was used to parse the HTML code to extract the only the relevant content. At this initial stage the built in HTML parser in python was used. However, on a preliminary analysis, it was observed that there remained a large amount of unnecessary text, including texts for advertisements, links to other articles, code for adblockers etc. which did not form a part of the main article. Since the presence of irrelevant text would affect the analysis, further cleaning was undertaken. The .find function from beautiful soup was used to extract the headline and article body for each article. This was done by extracting the contents of the relevant div through the class id. Since the HTML code for every publication varied, the class id for the article body for each publication differed. Therefore, to find the relevant class id’s specific to each publication, the HTML code for each publication was inspected. Using a for loop and if conditions, relevant headlines and article contents were then stored in a separate column in the data frame. 

Analysis of the comments section would have provided additional data on the perception of the public on the issue of congestion pricing. However, NY Daily did not have a visible comments section, and the Gothamist had very limited comments for its articles. Therefore, in the absence of a robust comments section for all three of the selected publication, this data was not analyzed for the purposes of this project. 

####ii.	Removal of stop words: 
In order to ensure that punctuation and common words used in the English were not included in the analysis, the NLTK library was imported and stop words were excluded from the list of words that would be analyzed. In addition to the stop words in the NLTK library certain other irrelevant words were also included to the list of stop words such as make, go, even etc to aid in more accurate analysis. 

####iii.	Lemmatization: 
The words were also reduced down to their base form using the .lemmatize function. As a result different forms of the same word were grouped together, and the number of unique words were thus reduced. This aided in more accurate analysis of patterns, i.e., words used frequently in the selected articles.

Note: 
Step 1, was undertaken to carry out sentiment analysis. 
Steps 2 and 3 were undertaken for the process of creating word clouds in an effort to analyze the words frequently used by different publications. 

###C.	Text Analysis 

####i.	Sentiment Analysis: 
In order to identify the perception of the public towards the introduction of congestion pricing in Manhattan, a sentiment analysis of the selected 30 articles were undertaken. The headlines and the articles were analyzed to generate a sentiment score through Vader. Each article was given a composite score that ranged from -1 to 1. A score of -1 to 0 would be considered as having a negative tone towards congestion pricing. A score from 0 to 1 would be considered as having a positive tone towards congestion pricing. A score of 0 would be considered neutral. The composite sentiment score for each article was then added as a separate column in the data frame. 

####ii.	Word Clouds: 
In addition to conducting a sentiment analysis, the frequency of words used by different publications were also analyzed to gauge whether there were certain words that were used more frequently by specific publications. This was done in an effort to determine if there were any particular issues related to congestion pricing that were being covered be each publication. 4 separate word clouds were generated, one for each publication and a comprehensive one for all three articles.

```python
#To cull out the headline and content from the HTML code by creating a for loop and add it to a column in the dataframe as well as include it in seprate .txt files
#Code to extract article body was adapted from various tutorials/articles including:
#https://www.geeksforgeeks.org/how-to-extract-a-div-tag-and-its-contents-by-id-with-beautifulsoup/
#https://www.pluralsight.com/guides/extracting-data-html-beautifulsoup
#HTML markers of div and class name were identified by inspecting the HTML source code for the URLs of each publication
id = 0
data_df['article_text']=""

for i, row in data_df.iterrows():
    soup = BeautifulSoup(row['text'], 'html.parser')
    if i < 10:
        article_content = soup.find('div', {'class': 'entry-content'})
        headline = soup.find('h1').get_text()
    elif i >= 10 and i <= 20:
        article_content = soup.find('div', {'class': 'article-body'})
        headline = soup.find('h1').get_text()
    elif i >= 21 and i <= 30:
        article_content = soup.find('div', {'class': 'col-sm-md-12 col-lg-xl-8 left-article-section ie-flex-100-percent-sm layout-section right-article-section'})
        headline = soup.find('h1').get_text()

    print(f"Article {id}: {headline}")

    id += 1
    data_df.at[i, 'article_text'] = data_df.at[1,'Headline'] + '\n\n'+ article_content.get_text() + '\n'
    with open(f"cp_articles/article_{id}.txt", "w") as file:
            file.write(f"{headline}\n\n")
            if article_content:
                file.write(article_content.get_text() + '\n')
    
```

```python
#Running a sentiment analysis and then adding a column to the dataframe to contain sentiment scores of each article
#adapted from class curiculum
analyser = SentimentIntensityAnalyzer()
text_col = 'article_text'
def get_sentiment(row):
    return analyser.polarity_scores(row[text_col])['compound']
data_df['sentiment'] = data_df.apply(get_sentiment, axis=1)
```

```python
# group sentiment scores by article index
grouped = data_df.groupby(data_df.index)['sentiment'].mean()

# define colors based on index range
colors = ['red' if i < 10 else 'grey' if i < 20 else 'blue' for i in range(30)]

# create bar chart with colored bars
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(range(1, 31), grouped, color=colors, width=0.7)
ax.set_xlabel('Article Number', fontweight='bold')
ax.set_ylabel('Compound Sentiment Score', fontweight='bold')
ax.set_title('Sentiment Scores by Article', fontweight='bold')

# set x axis ticks and limits
ax.set_xticks(range(1, 31))
ax.set_xlim([0.5, 30.5])

plt.show()
```

```python
#Importing necessary libraries for data cleaning and adding to the stopwords list
import nltk

from nltk.corpus import stopwords
stops = stopwords.words('english')
stops.extend(['see', 'already', 'go', 'take','make','even' ])

from nltk import pos_tag
from nltk.corpus import wordnet 
from nltk.stem.wordnet import WordNetLemmatizer

from wordcloud import WordCloud    
import matplotlib.pyplot as plt

import glob
```

```python
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

```python
#To clean the data by culling out the words, lemmatizing them and removing stop words
#Cleaned data is then segregated into 4 collections:master group and by publication
id = 0
filepaths = []
all_text=[]
nypost=[]
gothamist=[]
nydaily=[]


for file in cp_articles:
    filepaths.append(file)
for filepath in filepaths:
    text = open(filepath, encoding=encoding).read()
    text_tokens = text.split()
    text_lower = [t.lower() for t in text_tokens if t.isalnum()]

    text_stops = [t for t in text_lower if t not in stops]

    text_clean = [WordNetLemmatizer().lemmatize(t, get_wordnet_pos(t)) for t in text_stops]
    all_text += text_clean
    
    unique_string = ' '.join(text_clean)
 
        
    with open(f"cp_articles_cleaned/article_cleaned_{id}.txt", "w") as file:
        file.write(unique_string)
    all_text.extend(text_clean)
    if id<10:
        nypost.extend(text_clean)
    elif id>=10 and id<=20:
        gothamist.extend(text_clean)
    elif id>=21 and id<=30:
        nydaily.extend(text_clean)
    id += 1
    
unique_string = ' '.join(all_text)
nypost_string = ' '.join(nypost)
gothamist_string = ' '.join(gothamist)
nydaily_string = ' '.join(nydaily)
  
```

```python
#Generating 4 separate word clouds and formatting them for font, background, size etc
font_path = "/Library/Fonts/Arial.ttf"
wordcloud_all = WordCloud(max_font_size=70, background_color='white', max_words=40, font_path=font_path).generate(unique_string)
wordcloud_nypost = WordCloud(max_font_size=40, background_color='white', max_words=20, font_path=font_path).generate(nypost_string)
wordcloud_gothamist = WordCloud(max_font_size=40, background_color='white', max_words=20, font_path=font_path).generate(gothamist_string)
wordcloud_nydaily = WordCloud(max_font_size=40, background_color='white', max_words=20, font_path=font_path).generate(nydaily_string) 
```

```python
#defining colour schemes for the four word clouds
#Colour codes were determined through the following link: https://matplotlib.org/stable/tutorials/colors/colormaps.html
color_map_all = plt.cm.get_cmap('rainbow')
def color_func1(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(int(255*x) for x in color_map_all(np.random.rand()))

color_map_nypost = plt.cm.get_cmap('autumn')
def color_func2(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(int(255*x) for x in color_map_nypost(np.random.rand()))

color_map_gothamist = plt.cm.get_cmap('gist_gray')
def color_func3(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(int(255*x) for x in color_map_gothamist(np.random.rand()))

color_map_nydaily = plt.cm.get_cmap('winter')
def color_func4(word, font_size, position, orientation, random_state=None, **kwargs):
    return tuple(int(255*x) for x in color_map_nydaily(np.random.rand()))
```

```python
#generating 4 separate word clouds and then formatting them
#Adapted from various tutorials including: 
#https://www.datacamp.com/tutorial/wordcloud-python
#https://medium.com/mlearning-ai/wordclouds-with-python-c287887acc8b

wordcloud_all.recolor(color_func=color_func1)
plt.figure()
plt.imshow(wordcloud_all, interpolation="bilinear")
plt.axis("off")
plt.title('All Articles',fontsize=20,color='Black', y=1.1)
plt.text(x=0, y=-10, s='This word cloud shows the most frequent words in all articles analyzed.',fontstyle='italic', fontsize=12, color='Black', horizontalalignment='left')
wordcloud_all.to_file("wordcloud_all.png")
plt.show()

wordcloud_nypost.recolor(color_func=color_func2)
plt.figure()
plt.imshow(wordcloud_nypost, interpolation="bilinear")
plt.axis("off")
plt.title('NY Post',fontsize=20,color='Black', y=1.1)
plt.text(x=0, y=-10, s='This word cloud shows the most frequent words in the NY Post articles analysed.',fontstyle='italic', fontsize=12, color='Black', horizontalalignment='left')
wordcloud_nypost.to_file("wordcloud_nypost.png")
plt.show()

wordcloud_gothamist.recolor(color_func=color_func3)
plt.figure()
plt.imshow(wordcloud_gothamist, interpolation="bilinear")
plt.axis("off")
plt.title('Gothamist', fontsize=20,color='Black', y=1.1)
plt.text(x=0, y=-10, s='This word cloud shows the most frequent words in the Gothamist articles analysed.',fontstyle='italic', fontsize=12, color='Black', horizontalalignment='left')
wordcloud_gothamist.to_file("wordcloud_gothamist.png")
plt.show()

wordcloud_nydaily.recolor(color_func=color_func4)
plt.figure()
plt.imshow(wordcloud_nydaily, interpolation="bilinear")
plt.axis("off")
plt.title('NY Daily', fontsize=20,color='Black', y=1.1)
plt.text(x=0, y=-10, s='This word cloud shows the most frequent words in the NY Daily articles analysed.',fontstyle='italic', fontsize=12, color='Black', horizontalalignment='left')
plt.show()

wordcloud_all.to_file(f"wordclouds/word_cloud_all.png")
wordcloud_nypost.to_file(f"wordclouds/word_cloud_nypost.png")
wordcloud_gothamist.to_file(f"wordclouds/word_cloud_gothamist.png")
wordcloud_nydaily.to_file(f"wordclouds/word_cloud_nydaily.png")

```
